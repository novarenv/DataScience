## 4. Masukkan nilai X dan Y

```python
x = data[['panjang kelopak', 'lebar kelopak', 'panjang mahkota', 'lebar mahkota']]
y = data['spesies']

knn = KNeighborsClassifier()
```

#### Parameter/argumen pada algoritma/<i>class</i> K-Nearest Neighbor

```python
knn
```

algorithm = <a href="https://en.wikipedia.org/wiki/Ball_tree">ball_tree</a>, <a href="https://en.wikipedia.org/wiki/K-d_tree">kd_tree</a>, <a href="https://en.wikipedia.org/wiki/Brute-force_attack">brute</a>, auto<br>
leaf_size = Akan berfungsi apabila ball_tree atau kd_tree dipilih. Mempengaruhi kecepatan query data(tergantung memory juga)<br>
metric = https://en.wikipedia.org/wiki/Minkowski_distance<br>
metric_params = dict<br>
n_jobs = int<br>
n_neighbors = int<br>
p = 1 : rumus <i>manhattan distance</i>, 2 : rumus <i>euclidean distance</i>(<i>default</i>)<br>
weights = str/callable, uniform(<i>default</i>)<br>

## 5. <i>Fit</i> dan cek akurasi

```python
knn.fit(x, y)
knn.score(x,y)
```

fit() = Digunakan untuk membuat sebuah model pembelajaran mesin. Jadi si <b><i>machine</i></b> itu <b><i>learning</i></b>'nya ya lewat fungsi ini<br>
score() = Digunakan untuk menguji akurasi model
